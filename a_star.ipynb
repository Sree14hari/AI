```python
import heapq
class node:
    def __init__(self,name,heuristic):
        self.name=name
        self.heuristic=heuristic
        self.neighbours={}
        self.g=float("inf")
        self.parent=None
    
    def addNeighbour(self,neighbour,cost):
        self.neighbours[neighbour]=cost
    
    def __lt__(self,other):
        return (self.g+self.heuristic)<(other.g+other.heuristic)


def Astar(start,goal):
    open_list=[]
    closed_list=set()
    start.g=0
    heapq.heappush(open_list,start)
    
    while open_list:
        current_node=heapq.heappop(open_list)
        print(f"Expanding node {current_node.name} | g={current_node.g} | h={current_node.heuristic}")
        
        if current_node.name in closed_list:
            continue
        
        if current_node.name==goal:
            return reconstruct(goal)
        
        closed_list.add(current_node.name)
        
        for neighbour , cost in current_node.neighbours.items():
            tent_g=current_node.g+cost
            if tent_g<neighbour.g:
                neighbour.parent=current_node
                neighbour.g=tent_g
                heapq.heappush(open_list,neighbour)
    
    print("Goal not found")   
    return None

def reconstruct(goal):
    path=[]
    current=goal
    while current:
        path.append(current.name)
        current=current.parents
    path.reverse()
    return path

if __name__=="__main__":
    a=node('A',5)
    b = node('B', 3)
    c = node('C', 4)
    d = node('D', 2)
    e = node('E', 0)   
    
```
